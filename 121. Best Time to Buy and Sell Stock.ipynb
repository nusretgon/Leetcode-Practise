```python
def maxProfit(prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        maxProfit = 0
        minBuy = prices[0]
        for i in range(len(prices)):
            if minBuy>prices[i]:     # We can use min function but I want to explain clearly
                minBuy=prices[i]
            if maxProfit<prices[i]-minBuy:  # Also max function could be used.
                maxProfit=prices[i]-minBuy
        return maxProfit
        
```

```python
prices = [7,1,5,3,6,4]
maxProfit(prices)
```

```python
prices = [7,6,4,3,1]
maxProfit(prices)
```

```python
prices = [8,10,9,3,6,9,1,4,6,7,8,0,4,2,1,5,6,7,8,9,0,6,4,7,8,9,7,4,3,5,7,8]
maxProfit(prices)
```
