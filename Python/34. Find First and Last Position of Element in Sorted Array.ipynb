```python
def searchRange(nums, target):
        list1=[]
        numberof=nums.count(target)
        if numberof>0:
            start=nums.index(target)
            list1.append(start)
            list1.append(start+numberof-1)
            return list1
        else:
            return [-1,-1]
```

```python
nums = [5,7,7,8,8,10]
target = 8
searchRange(nums,target)
```
