```python
s ="dvdf"
```

```python
def lengthOfLongestSubstring(s):
        """
        :type s: str
        :rtype: int
        """
        a=""
        maxCount=1
        if len(s)==0:
            return 0
        for i in s: 
            if (a.find(i)!=-1):
                if maxCount<len(a):
                    maxCount=len(a)
                a=a[a.find(i)+1:]        # To remove the duplicate
                a+=i                     
            else:
                a+=i
            if maxCount<len(a):
                maxCount=len(a)
            print(i,":",maxCount)        # To see the progress
        return maxCount
```

```python
lengthOfLongestSubstring(s)
```
