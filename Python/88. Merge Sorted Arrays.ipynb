```python
def merge(nums1, m, nums2, n):
        """
        :type nums1: List[int]
        :type m: int
        :type nums2: List[int]
        :type n: int
        :rtype: None Do not return anything, modify nums1 in-place instead.
        """
        length=len(nums1)
        while length>m:
            nums1.pop()
            length-=1
        while nums2:
            nums1.insert(m,nums2.pop())
            m+=1
        nums1.sort()
        return nums1                  # Not needed but I show.
```

nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3

```python
merge([1,2,3,0,0,0],3,[2,5,6],3)
```

```python
merge([-1,-1,0,0,0,0],4,[1,2,3],3)
```

```python
merge([1,2,3],3,[-1,-1,0,0],4)
```

```python
def merge1(nums1, m, nums2, n):      # Creepy solution
    nums1[m:]=nums2
    nums1.sort()
    return nums1
```

```python
merge1([1,2,3,0,0,0],3,[2,5,6],3)
```
