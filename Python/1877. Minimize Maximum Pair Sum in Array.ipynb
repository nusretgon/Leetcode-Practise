```python
def minPairSum(nums):
    right=len(nums)-1
    left=0
    nums.sort()
    nums1=[]
    while right>left:
        nums1.append(nums[right]+nums[left])
        right-=1
        left+=1
    return max(nums1)
```

```python
nums = [3,5,4,2,4,6]
minPairSum(nums)
```

```python
nums = [3,4,4,2,4,7,1,2,3,4,5,6,7,8,9,0,4,2,1,5]
minPairSum(nums)
```
