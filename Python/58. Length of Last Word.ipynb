```python
def lengthOfLastWord(s):
    s=s.strip()
    leng=len(s)
    a=0
    for i in range(leng-1,-1,-1): 
        if s[i]!=" ":
            a+=1
        else:
            return a
    return a
```

```python
s="Nusret Gön"
lengthOfLastWord(s)
```

```python
s="asdmamdalk    aksmdamlsdka    "
lengthOfLastWord(s)
```

```python
def lengthOfLastWord(s):
        s=s.strip()
        s=s[::-1]
        a=0
        for i in s: 
            if i!=" ":
                a+=1
            else:
                return a
        return a
                
```
