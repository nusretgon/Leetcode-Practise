```python
def searchInsert(nums, target):
    r=len(nums)-1
    l=0
    while l<=r:
        index = (r+l)//2
        if nums[index]>target:
            r=index-1      
        elif nums[index]<target:
            l=index+1
        else:
            return index
    return l
```

```python
nums = [1,3,5,6]
target = 5
searchInsert(nums,target)
```
