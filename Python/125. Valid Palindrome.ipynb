```python
def isPalindrome(s):
        s=s.lower()
        for i in s:
            if i.isalpha()==False and i.isdigit()!=True:
                s=s.replace(i,"")
        return s==s[::-1]
```

```python
s = "A man, a plan, a canal: Panama"
isPalindrome(s)
```

```python
s="000P"
isPalindrome(s)
```

```python
def isPalindrome(s):
        """
        :type s: str
        :rtype: bool
        """
        s=s.lower()
        c=""
        for i in s:
            if i.isalpha()==True or i.isdigit()==True:
                c+=i
        return c==c[::-1]
```

```python
s = "A man, a plan, a canal: Panama"
isPalindrome(s)
```

```python
s="000P"
isPalindrome(s)
```
